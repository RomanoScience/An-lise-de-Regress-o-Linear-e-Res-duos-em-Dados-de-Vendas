```python
# Será utilizado tecnica dos minimos quadrados ou regressão linear para analisar a relação entre duas variaveis como tempo (x) e o salario (y)

# 1° Importar biblioteca de manipulação dos dados em grupo ou tabelas
# 2° Importar bibliteoca de vizualização gráfica em formato de dados ou funções
import pandas as pd
import matplotlib_inline as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
# 3° carregar o arquivo usando a biblioteca panda usando pd e indicar o caminho para carregar o arquivo
df = pd.read_csv ("C:/Users/ronal/Desktop/Analise_anomalias/Data/vendas_ficticias.csv", sep = ",")
# 4° Verificar quais são os tipos de dados contigo do data frame df
verificar = [ df.info]
# 5 ° inserir os nomes das colunas independentes que será traçadas a funcao no formato y = ax + b
x = df[["Quantidade"]]
y = df[["Preço Unitário"]]
modelo = LinearRegression()
modelo.fit(x,y)
# Comando para fazer a imagem da função com base nos dados em x, y é a imagem da função aproximada
imagem = modelo.predict(x)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
# 6° Plotando os dados reais 
axs[0].scatter(x, y, color='blue')
# 7° Plotando a reta de regressão
axs[0].plot(x, imagem, color='red')
axs[0].set_title('Regressão Linear')
axs[0].set_xlabel('Quantidade')
axs[0].set_ylabel('Preço unitário')

# 8° Calculando o Desvio padrão para avaliar se os dados estão muito dispersos da média ou não
desvio_padrao = np.std(imagem)
print("Desvio padrão é: ",desvio_padrao)
# 9° Calculo do desvio padrão residual é um medida para avaliar se a curva traçada representa bem o modelo e os dados estarão proximo a ela
erro = (y - imagem)
desvio_residual = np.std(erro)
axs[1].scatter(imagem, erro, color='blue')
axs[1].axhline(y=0, color='black', linestyle='--')
axs[1].set_title('Resíduos do Modelo')
axs[1].set_xlabel('Quantidade')
axs[1].set_ylabel('Resíduo')

plt.tight_layout()
plt.show()
```
